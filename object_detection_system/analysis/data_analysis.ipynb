# Quality and resolution of Input Frames

- High quality images with clear, sharp details and appropriate resolutions are important for the model to accurately detect and classify objects. Low-resolution or blurry images can lead to missed detections or incorrect classifications.
- Poor image quality can then degrade the performance of the model by reducing its ability to identify features necessary for accurate predictions. This can lead to a higher rate of false positives and false negatives.
- An image captured in low light or with motion blur might cause the model to fail to detect a sign or a car. 




# Variability in Object Appearance and Context:
- Objects in real-world scenarios can appear in various shapes, sizes, colors, and orientations. Additionally, the context or background where objects appear can vary widely.
- High variability can challenge the model's generalization capabilities. If the training data does not cover a wide range of variations, the model may struggle to detect and classify objects in un trained scenarios.

# Occlussions and Overlapping Objects
- We can call an occlussion when and object is partially hidden by another object, while overlapping happens when objects are close to each other or touching. These situations can complicate the detection process.
- The model may find it difficult to correctly identify and separate objects that are occluded or overlappping. This can lead to inaccurate bounding boxes or misclassification of objects.
